# PCR Solution

```python
import os

primer_f = "TCCCAAGCGTAACGATTCAG"
primer_r = "TTTGGCAGCCGTCATAATCC"


def read_file(filepath):
    with open(filepath) as f:
        file_content = []
        for line in f:
            if not line.startswith(">"):
                file_content.append(line.strip())
        return "".join(file_content)


def rev_comp(primer):
    comps = {"T": "A", "C": "G", "A": "T", "G": "C"}
    rev_comp = [comps[i] for i in primer][::-1]
    return "".join(rev_comp)


def pcr(seq, primer_f, primer_r):
    primer_r = rev_comp(primer_r)
    pos_f = seq.find(primer_f) + 1
    pos_r = seq.find(primer_r) + 20
    product_size = pos_r - pos_f + 1
    return product_size
    
    
def main():
    folder = "files"
    out_file = "answers.csv"
    num_of_files = 0
    for filename in os.listdir(folder):
        if filename.endswith(".fasta"):
            num_of_files += 1
    amplicon_list = []
    for filenum in range(num_of_files):
        filepath = os.path.join("files", str(filenum + 1) + ".fasta")
        seq = read_file(filepath)
        amplicon_size = pcr(seq, primer_f, primer_r)
        print(filepath, amplicon_size)
        amplicon_list.append(amplicon_size)
    with open(out_file, "w") as f:
        for i in amplicon_list:
            f.write("{}\n".format(i))
    print("done")
    
    
main()
```
